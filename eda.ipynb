```python
import numpy as np
import matplotlib.pyplot as plt

from src.utils import load_data, get_data_target_field_names
```

```python
datasets = load_data()
for dataset in datasets:
    print(len(dataset))
```

```python
targets = {}
for i, dataset in enumerate(datasets):
    targets[i] = []
    for mol in dataset:
        targets[i].append(mol.GetProp(get_data_target_field_names()[i]))
    targets[i] = np.array(targets[i]).astype(np.float64)
```

```python
fig, ax = plt.subplots(1, 5, figsize=(20, 6))
for i in range(5):
    ax[i].hist(targets[i])
    ax[i].set_title(get_data_target_field_names()[i])
    ax[i].set_xlabel(get_data_target_field_names()[i])
    ax[i].set_ylabel('count')
    ax[i].set_yscale('log')
plt.plot()
```
